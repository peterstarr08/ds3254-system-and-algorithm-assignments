```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def On2(x, a,b):
    return a * np.square(x) + b
```

```python
df = pd.read_csv('sort_data.txt')

baseInstructionSpeed = 2.6e9 # change according to your PC

xValue = df['size']
insertion_sort_value = df['time_insert_sort']* baseInstructionSpeed
selection_sort_value = df['time_selection_sort']* baseInstructionSpeed
merge_sort_value = df['time_merge_sort']* baseInstructionSpeed

popt, pcov = curve_fit(On2, xValue, insertion_sort_value)
```

```python
plt.plot(xValue, insertion_sort_value, label='Insertion Sort', color='b')
plt.plot(xValue, On2(xValue, *popt), label='Fitted Curve', color='g')
plt.plot(xValue, np.square(xValue), label='O(n^2)', color='r')
plt.show()
```

```python
popt, pcov = curve_fit(On2, xValue, selection_sort_value)
plt.plot(xValue, selection_sort_value, label='Selection Sort', color='b')
plt.plot(xValue, On2(xValue, *popt), label='Fitted Curve', color='g')
plt.plot(xValue, np.square(xValue), label='O(n^2)', color='r')
plt.show()
```

```python
def OnLogn(x, a, b):
    return a * x * np.log2(x) + b
```

```python
popt, pcov = curve_fit(OnLogn, xValue, merge_sort_value)
plt.plot(xValue, merge_sort_value, label='Merge Sort', color='b')
plt.plot(xValue, OnLogn(xValue, *popt), label='Fitted Curve', color='g')
plt.plot(xValue, xValue*np.log2(xValue), label='O(n log(n))', color='r')
plt.show()
```
